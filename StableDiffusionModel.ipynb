```python
!pip install diffusers
!pip install transformers
!pip install gradio==3.48.0


import gradio as gr
from diffusers import StableDiffusionPipeline
import torch


model_id="runwayml/stable-diffusion-v1-5"
sd_pipeline= StableDiffusionPipeline.from_pretrained(model_id,torch_dtype=torch.float16)
sd_pipeline= sd_pipeline.to("cuda")


def get_completion(prompt):
  negative_prompt="""
  complex background, low quality , bad anatomy ,watermark,signature, text, extra digit, fewer digits, cropped, worst quality, low quality, normal quality, jpeg artifacts, signature, watermark, username, blurry
  """
  return sd_pipeline(prompt=prompt,negative_prompt=negative_prompt).images[0]


  genai_app =  gr.Interface(fn=get_completion,
                            inputs=gr.components.Textbox(Label="Input prompt"),
                            outputs=[gr.Image(label="Result")],
                            title="Generate cool images from your prompts",
                            Description="Generate any image with stable diffusion",
                            allow_flagging="never",
                            examples=["astranaut, riding a horse on Mars","cat listening Caravan"])

genai_app.launch()





```
